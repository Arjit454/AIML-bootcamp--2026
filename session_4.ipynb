```python
import pandas as pd
import numpy as np

# Read CSV and split the comma-separated values
df = pd.read_csv('studentrecord.csv', sep=',')

# If it still loads as one column, do this:
df = df.iloc[:, 0].str.split(',', expand=True)

# Set column names
df.columns = [
    'student_id', 'name', 'age', 'city', 'math', 'science', 'english',
    'attendance_percent', 'study_hours_per_week',
    'assignments_submitted', 'final_project_score'
]

# Convert math column to NumPy array
math_scores = df['math'].astype(int).to_numpy()

print(math_scores)
```

```python
math_scores = df['math'].to_numpy()

print(math_scores)
```

```python
print(type(math_scores))
```

```python
import numpy as np
mean_math = math_scores.mean()
print(mean_math)
```

```python
df['avg_score'] = (
    df['math'].astype(float) +
    df['science'].astype(float) +
    df['english'].astype(float)
) / 3
```

```python
tstudents = df[df['avg_score'] > 85]
print(tstudents[['name', 'avg_score']])
```

```python
lattendance = df[df['attendance_percent'].astype(float) < 75]
print(lattendance[['name', 'attendance_percent']])
```

```python


mathura_students = df[df["city"] == "Mathura"]

print(mathura_students)
```

```python
df["attendance_percent"] = df["attendance_percent"].astype(float)

low_attendance = df[df["attendance_percent"] < 75]

print(low_attendance)
```

```python
df["math"] = pd.to_numeric(df["math"])
df["science"] = pd.to_numeric(df["science"])
df["english"] = pd.to_numeric(df["english"])

df["average"] = (df["math"] + df["science"] + df["english"]) / 3

print(df[["name", "average"]])
```

```python
result = []

for avg in df["average"]:
    if avg >= 40:
        result.append("Pass")
    else:
        result.append("Fail")

df["result"] = result

print(df[["name", "average", "result"]])
```

```python
class_average = df["average"].mean()

print("Class Average =", class_average)
```

```python
highest_average = df["average"].max()

print("Highest Average =", highest_average)

lowest_average = df["average"].min()

print("Lowest Average =", lowest_average)
```

```python
highest_average = df["average"].max()

top_students = df[df["average"] == highest_average]

print(top_students[["name", "average"]])
```

```python
df["attendance_percent"] = pd.to_numeric(df["attendance_percent"])

low_attendance = df[df["attendance_percent"] < 75]

print(low_attendance[["name", "attendance_percent"]])
```

```python
df["assignments_submitted"] = pd.to_numeric(df["assignments_submitted"])

few_assignments = df[df["assignments_submitted"] < 6]

print(few_assignments[["name", "assignments_submitted"]])
```

```python
df.to_csv("session4_student_report.csv", index=False)

print("File Saved Successfully")
```
